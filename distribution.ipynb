```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
'''离散型分布：二项分布'''
x1 = np.random.binomial(n=10,p=0.3,size=100)    
#x1是每10次实验中成功的次数，这里的x1中有100个值
#进行100组实验，每组10次，每次成功的概率0.3，记录下每组成功的次数。
print(np.mean(x1))    #numpy计算均值
print(np.ptp(x1))    #极差
print(np.var(x1))    #方差
print(stats.mode(x1))    #numpy没有众数，从stats中计算众数

plt.hist(x1,bins=10,color='g',alpha=0.3)    #绘制x1的分布直方图
plt.xlabel('number')
plt.ylabel('frequency')
plt.title('distribution histogram')
plt.show()


#绘制任一组实验的x2概率分布图
n=10
p=0.3
x2 = np.arange(0,n+1,1)    #x2是10次实验中的所有可能取值
y = stats.binom.pmf(x2,10,0.3)    #y是概率分布函数
plt.plot(x2,y,marker='.')
plt.xlabel('random variable : x2')
plt.ylabel('probability : y')
plt.title('binomial distribution')
plt.grid()
plt.show()
```
